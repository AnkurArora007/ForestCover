```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import norm
plt.rcParams['figure.figsize']=(12,8)
%matplotlib inline
```

```python
df = pd.read_csv('covtype.csv')
print('Dataset has {} rows and {} columns'.format(df.shape[0], df.shape[1]))
df.head()
```

```python
df['Cover_Type'].value_counts().sort_index().plot.bar()
```

```python
sns.boxplot(x=df['Cover_Type'], y=df['Elevation'])
```

```python
sns.boxplot(df['Cover_Type'], df['Aspect'])
```

```python
for i in range(1,5):
    col = 'Wilderness_Area'+str(i)
    (df[col].groupby(df['Cover_Type']).sum()*100/len(df[df[col] == 1])).plot.bar()
    plt.title('%age of trees covered in '+col)
    plt.show()
```

```python
for i in range(1,41):
    col = 'Soil_Type'+str(i)
    (df[col].groupby(df['Cover_Type']).sum()*100/len(df[df[col] == 1])).plot.bar()
    plt.title('%age of trees grown in '+col)
    plt.show()
```
